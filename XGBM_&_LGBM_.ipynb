**Build and Evaluate XGBoost Model**

```python
# Predict and evaluate
y_pred_xgb = xgb_model.predict(X_val)
xgb_accuracy = accuracy_score(y_val, y_pred_xgb)
xgb_precision = precision_score(y_val, y_pred_xgb)
xgb_recall = recall_score(y_val, y_pred_xgb)
xgb_f1 = f1_score(y_val, y_pred_xgb)

print(f"XGBoost - Accuracy: {xgb_accuracy:.4f}, Precision: {xgb_precision:.4f}, Recall: {xgb_recall:.4f}, F1-score: {xgb_f1:.4f}")
```

**Strengths and Weaknesses**

**LightGBM:** Has a slightly higher accuracy and F1-score compared to XGBoost in both initial and tuned versions.

**XGBoost:** Shows slight improvement in precision and recall after tuning, but still lags behind LightGBM in overall performance metrics.

From the results, LightGBM appears to perform slightly better overall compared to XGBoost, especially in terms of accuracy and F1-score. However, the differences are relatively small, and both models show good performance.

These insights can help in understanding the trade-offs between using LightGBM and XGBoost, and further tuning or ensembling methods might be explored to boost the performance of the predictive model.
